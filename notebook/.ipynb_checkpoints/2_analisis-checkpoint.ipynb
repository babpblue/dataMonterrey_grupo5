# Analisis unidimensional variable Fecha

```python
compras_df['FECH EMI'].value_counts()
```

```python
# Validamos y contamos por mes
compras_df['FECH EMI'] = pd.to_datetime(compras_df['FECH EMI'], errors='coerce')

```

```python
conteo_por_mes = compras_df.groupby(pd.Grouper(key='FECH EMI', freq='M')).size()
conteo_por_mes
```

```python
conteo_por_mes.values
```

```python
conteo_por_mes.plot(kind='hist', bins=10, edgecolor='black')
```

```python
# Validamos y contamos por día
conteo_por_dia = compras_df.groupby(pd.Grouper(key='FECH EMI', freq='D')).size()
conteo_por_dia
```

```python
conteo_por_dia.values
```

```python
conteo_por_dia.plot(kind='hist', bins=10, edgecolor='black')
```

## Analisis Muldimensional de las variables

```python
compras_df.dtypes
```

```python
compras_df.head()
```

```python
sum_imoprte_beneficiario = compras_df.groupby('BENEFICIARIO')['IMPORTE'].sum()
```

```python
sum_imoprte_beneficiario.head(5)
```

```python
sum_imoprte_beneficiario.reset_index().sort_values(by='IMPORTE',ascending=False).iloc[0:10].plot(kind='bar', x='BENEFICIARIO',y='IMPORTE')
```

```python
# ii. Determinar el gasto promedio por cada categoría o por hoja de Excel.
```

```python
mean_importe_sheet =  compras_df.groupby('SHEET')['IMPORTE'].mean()
```

```python
mean_importe_sheet
```

```python
mean_importe_sheet.reset_index().sort_values(by='IMPORTE',ascending=False).iloc[0:10].plot(kind='bar', x='SHEET',y='IMPORTE')
```

```python
# iii. Determinar los gastos por fecha.
```

```python
sum_importe_fecha = compras_df.groupby('FECH EMI').sum()['IMPORTE']
```

```python
sum_importe_fecha.reset_index().sort_values(by='IMPORTE',ascending=False).iloc[0:10].plot(kind='bar', x='FECH EMI',y='IMPORTE')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
texto_compras = ''
```

```python
for row in compras_df['DESCRIPCION']:
    texto_compras = texto_compras + ' ' + row
```

```python
stopwords_list = stopwords.words('spanish') + ['2017','cont','millar','dpp']
```

```python
textos_compras = ' '.join(x for x in texto_compras.strip().lower().split(' ') if x not in stopwords_list and len(x)>4  )
```

```python
wc = WordCloud(background_color = 'white')
wc.generate(textos_compras)
```

```python
plt.axis("off")
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
